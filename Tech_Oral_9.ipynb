```python
# Over Sampling
```

```python
from imblearn.over_sampling import RandomOverSampler
ro=RandomOverSampler()
x_data,y_data=ro.fit_resample(x,y)
```

```python
from collections import Counter
print("Actual data",Counter(y))
print("Artificial data",Counter(y_data))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
r1=LogisticRegression()
r1.fit(x_train,y_train)
```

```python
y_pred=r1.predict(x_test)
y_pred
```

```python
y_pred=r1.predict([[40,1,1,140,289,0,1,172,0,0.0,2]])
y_pred
```

```python
ab=r1.score(x_test,y_test)*100
ab
```

```python
# Under sampling
```

```python
from imblearn.under_sampling import NearMiss
nm=NearMiss()
x1_data,y1_data=nm.fit_resample(x,y)
```

```python
from collections import Counter
print("Actual data",Counter(y))
print("Artificial data",Counter(y1_data))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1_data,y1_data,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
r1=LogisticRegression()
r1.fit(x_train,y_train)
```

```python
y_pred=r1.predict(x_test)
y_pred
```

```python
ad=r1.score(x_test,y_test)*100
ad
```

```python
# SMOTE
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
AA=x2_data,y2_data=sm.fit_resample(x,y)
```

```python
from collections import Counter
print("Actual data",Counter(y))
print("Artificial data",Counter(y2_data))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x2_data,y2_data,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
r1=LogisticRegression()
r1.fit(x_train,y_train)
```

```python
y_pred=r1.predict(x_test)
y_pred
```

```python
h=r1.score(x_test,y_test)*100
h
```
