```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
```

```python
data=pd.read_csv(r"D:\Machine learning dataset\Social_Network_Ads.csv")
data
```

```python
data["Purchased"].value_counts()
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=4,random_state=10, shuffle=True)
kf.get_n_splits(x)
print(kf)
```

```python
from sklearn.linear_model import LogisticRegression
l1=LogisticRegression()
```

```python
for train_index,test_index in kf.split(x):
    print("TRAIN:", train_index, "TEST:", test_index)
    x_train, x_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
    score=cross_val_score(l1,x_train,y_train,cv=kf)
    print("Score: ",score)
    print("Mean: ",np.mean(score)*100)
    y_pred=cross_val_predict(l1,x_test,y_test,cv=kf)
    print("Prediction: ",y_pred)
    (unique,counts)=np.unique(y_test,return_counts=True)
    frequencies=np.asarray((unique,counts)).T
    print(frequencies)
```

```python
from sklearn.linear_model import LogisticRegression
l1=LogisticRegression()
from sklearn.model_selection import cross_val_score
score=cross_val_score(l1,x_train,y_train,cv=kf)
print(score)
print(np.mean(score)*100)
```

```python
from sklearn.model_selection import cross_val_predict
y_pred=cross_val_predict(l1,x_test,y_test,cv=kf)
print(y_pred)
```
